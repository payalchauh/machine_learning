## Step 3 Seperate X (Predictor) and Y (Target features)

```python
# Make sure you use double square bracket for getting dataframe output
X = df[["RND"]]
Y = df[["PROFIT"]]
```

```python
X.head()
```

```python
Y.head()
```

### Step 4 Create a Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
model.intercept_
```

```python
model.coef_
```

### PROFIT_pred = 49032.9 + 0.8543 x RND
1. IF RND increase by 1 USD the PROFIT will increase by 0.8543 approximately
2. IF RND increase by 1000 USD then PROFIT will increase by 854.3 USD Approximately

```python
### Step 5 Model Evaluation
```

```python
# R2 score
model.score(X, Y)
```

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score,
)
```

```python
ypred = model.predict(X)
ypred[0:5]
```

```python
Y.head()
```

```python
def evaluate_model(model, x, y):
    # Predict the result on x
    ypred = model.predict(x)

    # Calculate different metrics
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)

    # Print the metrics
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.2%}")
    print(f"R2 : {r2:.2%}")
```

```python
evaluate_model(model, X, Y)
```

### R2 Score >= 0.8 hence above model is a good model
Hence i can use this model for out of sample prediction

### Step 6 Out of Sample Prediction
